```python
import os
import pandas as pd

# Define root directory containing patient folders
root_dir = "/home/sncr0/data/d1namo/diabetes_subset_pictures-glucose-food-insulin/diabetes_subset_pictures-glucose-food-insulin"

# Get all patient directories
patient_folders = sorted([f for f in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, f))])

# Iterate over all patient folders and read insulin.csv
for patient_folder in patient_folders:
    insulin_path = os.path.join(root_dir, patient_folder, "glucose.csv")

    # Check if insulin.csv exists
    if os.path.exists(insulin_path):
        df = pd.read_csv(insulin_path, memory_map=True)  # Read with memory mapping
        print(f"Patient {patient_folder}: {len(df)} records")
    else:
        print(f"Patient {patient_folder}: insulin.csv not found")
```
